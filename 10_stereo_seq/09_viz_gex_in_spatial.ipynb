```r
in_dir <- "../../results/10_stereo_seq/08_viz_cell_type_in_spatial"
```

```r
plot_marker <- function(obj, marker, pt.size.factor = 1.2,
                       max.cutoff = 'q99', min.cutoff = 'q1'){
    p <- SpatialFeaturePlot(obj, features = marker,
                            max.cutoff = max.cutoff, 
                            min.cutoff = min.cutoff,
                            pt.size.factor = pt.size.factor, 
                            image.alpha = 0) +
        scale_fill_viridis(option = 'A') +
        xlab("") + ylab("") +
        theme_cowplot() +
        theme(axis.text = element_blank(),
              axis.line = element_blank(),
              axis.ticks = element_blank(),
              legend.title = element_blank())
    
    
    return(p)


}
```

```r
obj <- readRDS(glue::glue("{in_dir}/HC01.rds"))
obj <- NormalizeData(obj)

options(repr.plot.height = 6, repr.plot.width = 12)

p1 <- plot_marker(obj, marker = "ACKR3", pt.size.factor = 1.2) + ggtitle("ACKR3")
p2 <- plot_marker(obj, marker = "SCARA5", pt.size.factor = 1.2) + ggtitle("SCARA5")

print(p1 + p2)
```
