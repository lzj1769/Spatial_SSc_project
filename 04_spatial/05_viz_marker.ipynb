```r
suppressMessages(library(Seurat))
suppressMessages(library(ggplot2))
suppressMessages(library(patchwork))
suppressMessages(library(dplyr))
suppressMessages(library(viridis))
suppressMessages(library(cowplot))
```

```r
input_dir <- "../../results/04_spatial/01_read_data"
```

```r
plot_marker <- function(obj, marker, pt.size.factor = 1.2){
    p <- SpatialFeaturePlot(obj, features = marker,
                            max.cutoff = "q99", 
                            min.cutoff = "q1",
                            pt.size.factor = pt.size.factor, 
                            image.alpha = 0) +
        scale_fill_viridis(option = 'A') +
        xlab("") + ylab("") +
        theme_cowplot() +
        theme(axis.text = element_blank(),
              axis.line = element_blank(),
              axis.ticks = element_blank())
    
    
    return(p)


}
```

```r
options(repr.plot.height = 4, repr.plot.width = 20)

obj <- readRDS(glue::glue("{input_dir}/HC01.rds"))

p1 <- plot_marker(obj, marker = 'DCN')
p2 <- plot_marker(obj, marker = 'KRT1')
p3 <- plot_marker(obj, marker = 'VWF')
p4 <- plot_marker(obj, marker = 'DCD')
p5 <- plot_marker(obj, marker = 'RGS5')

p <- p1 + p2 + p3 + p4 + p5 + plot_layout(ncol = 5)

print(p)
```
