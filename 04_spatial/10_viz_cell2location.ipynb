```r
input_dir <- "../../results/04_spatial/01_read_data"
out_dir <- "../../results/04_spatial/10_viz_cell2location"

if(!dir.exists(out_dir)){
    dir.create(out_dir, recursive = TRUE)
}
```

```r
sample_list <- c('HC01', 'HC02', 'HC03', 'HC05',
               #'SSc4733', 
                 'SSc4994', 
                 'SSc5380', 
                 #'SSc5722',
               'SSc-HL01', 'SSc-HL05', 'SSc-HL06', 'SSc-HL11',
               'SSc-HL13', 'SSc-HL25', 'SSc-HL33', 'SSc-HL35')
```

```r
for(sample in sample_list){
    obj <- readRDS(glue::glue("{input_dir}/{sample}.rds"))
    df <- read.csv(glue::glue('../../results/04_spatial/09_run_cell2location/{sample}.csv'), row.names = 1)
    obj <- AddMetaData(obj, metadata = df)
    
    saveRDS(obj, glue::glue("{out_dir}/{sample}.rds"))
}
```

```r
obj <- readRDS(glue::glue("{out_dir}/HC01.rds"))

cols <- ArchR::paletteDiscrete(obj@meta.data$pred_cell_type)

SpatialDimPlot(obj, group.by='pred_cell_type', 
               cols = cols, 
               pt.size.factor = 1.8, image.alpha = 0) + 
    guides(fill = guide_legend(override.aes = list(size = 5))) +
    ggtitle("HC01")
```
