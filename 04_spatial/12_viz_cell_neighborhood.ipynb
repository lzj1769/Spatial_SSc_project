```r
df <- read.csv(glue::glue("{in_dir}/HC01.csv"), row.names = 1)
    
ht <- Heatmap(as.matrix(df),
              name = "Enrichment",
              rect_gp = gpar(col = "black", lwd = 0.2),
              na_col = "white",
              column_title = "HC01")
    
draw(ht)
```
