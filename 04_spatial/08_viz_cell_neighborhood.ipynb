```r
df <- read.csv(glue::glue("../../results/04_spatial/07_cell_neighborhood/HC01.csv"), row.names = 1)
    
ht <- Heatmap(as.matrix(df),
              name = "Enrichment",
              rect_gp = gpar(col = "black", lwd = 0.2),
              na_col = "white",
              column_title = "HC01")
    
draw(ht)
```
