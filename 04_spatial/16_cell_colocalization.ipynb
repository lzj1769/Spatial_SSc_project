```r
suppressMessages(library(dplyr))
suppressMessages(library(ComplexHeatmap))
```

```r
df1 <- read.csv('../../results/04_spatial/13_cell_colocalization/healthy.csv') %>%
    subset(select = (-X))

df2 <- read.csv('../../results/04_spatial/13_cell_colocalization/ssc.csv') %>%
    subset(select = (-X))
```

```r
df <- rbind(df1, df2)
```

```r
nrow(df)
```

```r
df_cor <- cor(df, method = "spearman")
```

```r
options(repr.plot.height = 6, repr.plot.width = 7)

ht1 <- Heatmap(df_cor, 
               rect_gp = gpar(col = "black", lwd = 0.25),
                # cell_fun = function(i, j, x, y, width, height, fill) {
                #     grid.text(round(df_healthy[i, j], 2), x, y, gp = gpar(fontsize = 10))}
              )

draw(ht1)
```
