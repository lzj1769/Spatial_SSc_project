```python
import warnings
warnings.filterwarnings('ignore')

import scanpy as sc
import pandas as pd
import os
import anndata as ad
import scipy as sp
```

```python
adata = sc.read_h5ad('../../results/02_nc_2021_Tabib/scrna.integrated.clustered.h5ad')
```

```python
sc.set_figure_params(figsize=(12, 12))

sc.pl.umap(adata, color=['cell_type'], frameon=False, legend_loc='on data')
```

```python
adata.obs['cell_type'] = adata.obs['leiden']

# Keratinocytes
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b0\b', 'Keratinocytes', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b2\b', 'Keratinocytes', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b5\b', 'Keratinocytes', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b7\b', 'Keratinocytes', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b17\b', 'Keratinocytes', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b19\b', 'Keratinocytes', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b20\b', 'Keratinocytes', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b23\b', 'Keratinocytes', regex=True)

# Pericytes
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b9\b', 'Pericytes', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b11\b', 'Pericytes', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b13\b', 'Pericytes', regex=True)

# Endothelial
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b1\b', 'Endothelial', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b14\b', 'Endothelial', regex=True)

# Fibroblasts
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b3\b', 'Fibroblasts', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b4\b', 'Fibroblasts', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b6\b', 'Fibroblasts', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b12\b', 'Fibroblasts', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b21\b', 'Fibroblasts', regex=True)

# Neural
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b15\b', 'Neural', regex=True)

# Smooth Muscle
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b18\b', 'Smooth Muscle', regex=True)

# Mast cells
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b24\b', 'Mast cells', regex=True)

# T/NK
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b10\b', 'T cells', regex=True)
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b16\b', 'NK cells', regex=True)

# Melanocytes
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b25\b', 'Melanocytes', regex=True)

# Endothelial
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b26\b', 'Endothelial', regex=True)

# Secretory cells
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b22\b', 'Secretory cells', regex=True)

# Macrophage/DC
adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b8\b', 'Macrophages', regex=True)

adata.obs['cell_type'] = adata.obs['cell_type'].str.replace(r'\b27\b', 'cDC', regex=True)
```

```python
sc.pl.umap(adata, color=['cell_type'], frameon=False, legend_loc='on data')
```
