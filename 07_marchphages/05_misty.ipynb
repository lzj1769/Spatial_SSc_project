```python
import warnings

warnings.filterwarnings("ignore")

import os
import scanpy as sc
import decoupler as dc
import plotnine as p9
import liana as li
import pandas as pd

from liana.method import MistyData, genericMistyData, lrMistyData
from liana.method.sp import RandomForestModel, LinearModel, RobustLinearModel
```

```python
out_dir = '../../results/07_marchphages/05_misty'
os.makedirs(out_dir, exist_ok=True)
```

```python
sample_list = ['HC01', 'HC02', 'HC03', 'HC05',
               'SSc4994', 
               'SSc5380', 
                'SSc-HL01', 'SSc-HL05', 'SSc-HL06', 'SSc-HL11',
                'SSc-HL13', 'SSc-HL25', 'SSc-HL33', 'SSc-HL35']
```

```python
def run_misty(sample, out_dir):
    adata = sc.read_h5ad(f'../../results/spatial/filtered_data/{sample}.h5ad')
    
    adata.obsm['fib_cell_state_score'] = pd.read_csv(f'../../results/05_fibroblast/06_cell_state_score/{sample}_score.csv', index_col=0)
    adata.obsm['mac_cell_state_score'] = pd.read_csv(f'../../results/07_marchphages/04_cell_state_score/{sample}_score.csv', index_col=0)
    
    fib_cell_state_score = li.ut.obsm_to_adata(adata, 'fib_cell_state_score')
    mac_cell_state_score = li.ut.obsm_to_adata(adata, 'mac_cell_state_score')
    
    misty = genericMistyData(intra=fib_cell_state_score, 
                             extra=mac_cell_state_score, 
                             cutoff=0, bandwidth=100, n_neighs=6)
    
    misty(model=RandomForestModel, n_jobs=-1, verbose = True)
    misty.write_h5mu(f'{out_dir}/{sample}.h5mu')
    misty.uns['interactions'].to_csv(f'{out_dir}/{sample}.csv')
    
    return misty
```
