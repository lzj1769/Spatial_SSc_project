```r
suppressMessages(library(Seurat))
suppressMessages(library(ggplot2))
suppressMessages(library(patchwork))
suppressMessages(library(dplyr))
suppressMessages(library(viridis))
suppressMessages(library(cowplot))
suppressMessages(library(anndataR))
```

```r
plot_marker <- function(obj, marker, pt.size.factor = 1.2, 
                        max.cutoff = 'q99', min.cutoff = 'q1'){
    p <- SpatialFeaturePlot(obj, features = marker,
                            max.cutoff = max.cutoff, 
                            min.cutoff = min.cutoff,
                            pt.size.factor = pt.size.factor, 
                            image.alpha = 0) +
        scale_fill_viridis(option = 'D') +
        xlab("") + ylab("") +
        theme_cowplot() +
        theme(axis.text = element_blank(),
              axis.line = element_blank(),
              axis.ticks = element_blank(),
              legend.title = element_blank())
    
    
    return(p)


}
```

```r
in_dir <- "../../results/04_spatial/10_viz_cell2location"
```

```r
obj <- readRDS(glue::glue("{in_dir}/HC01.rds"))
```

```r
colnames(obj@meta.data)
```

```r
p1 <- plot_marker(obj, marker = "Macrophages", pt.size.factor = 2)
p2 <- plot_marker(obj, marker = "CD36", pt.size.factor = 2)
```
